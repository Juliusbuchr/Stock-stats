# Quick Stock Data Points

The purpose of this notebook is to show a quick summary of given ticker, that can be used to assess the stock's stats. 

```python
## Import libs 
import pandas as pd 
import yfinance as yf
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

```python
## Specifying the stock
stock = input("Stock Ticker: ")
```

```python
## Getting data from yfinance API
df = yf.download(stock, start="2017-01-01")
```

```python
## Getting daily returns
df["Returns"] = df["Adj Close"].pct_change()*100
df = df.dropna()
```

```python
## Descriptive Stats
df.describe()
```

```python
plt.hist(df["Returns"], bins=100, color='r', density=True, alpha=1)
plt.ylabel("Density")
plt.xlabel(f"{stock} Daily Returns pct.")
plt.title(f"{stock} Returns Distribution")
plt.show()
```

```python
plt.hist(df["Returns"], bins=100, color='g', density=True, alpha=1, cumulative=True)
plt.ylabel("Density")
plt.xlabel(f"{stock} Daily Returns pct.")
plt.title(f"{stock} Empirical CDF")
plt.show()
```

```python
plt.plot(df["Returns"], color='b')
plt.ylabel("Daily Return pct.")
plt.title(f"{stock} Daily Return")
plt.show()
```
